```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("First Spark application") \
    .getOrCreate()
```

⚡ **Diferencia clave**

*   Pandas:	Datos pequeños o medianos en tu PC
*   Spark (SparkSession):	Big Data, datos enormes distribuidos


Spark se usa cuando los datos no están en tu PC, sino en:

* Hadoop
* Databricks
* servidores en la nube (AWS, Azure, GCP)

Aquí SparkSession es la entrada para todo el sistema distribuido.

| Etapa del ML        | pandas         | Spark              |
| ------------------- | -------------- | ------------------ |
| Cargar datos        | ✔️             | ✔️                 |
| Limpiar datos       | ✔️             | ✔️ (si es grande)  |
| Feature engineering | ✔️             | ✔️ (a gran escala) |
| Entrenar modelo     | ✔️ (principal) | ⚠️ raro / limitado |
